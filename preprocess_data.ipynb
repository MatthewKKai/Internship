```python
import os
import cv2
from PIL import Image
import csv
```

```python
data_list = []
# data_list = [folder,file_path,file_name,height,wight,depth = 3]
def data_preprocessing(file_dir):
    temp = [] # 暂存变量表
    data_list = []
    for root, dir_names, file_names in os.walk(file_dir):
        if len(file_names)>0:
            for file_name in file_names:
                file_path = root+"/"+file_name
                img = cv2.imread(file_path)
                sp = img.shape
                height = sp[1]
                wight = sp[0]
                if "image" in root.split("/"):
                    folder = "image"
                    temp = [folder, file_path, file_name, height, wight, 3]
                elif "masks" in root.split("/"):
                    folder = "masks"
                    temp = [folder, file_path, file_name, height, wight, 3]
                else:
                    pass
                data_list.append(temp)
    return data_list
```

```python
file_dir = "/Users/loaferzk/LoaferZK/实习/深至/data/测试集"
data_list = data_preprocessing(file_dir)
```

```python
def dump_to_csv(csv_path, data_list):
    headers = ["folder","file_path","file_name","height","wight","depth"]
    with open(csv_path, "w", newline = "", encoding = "utf-8-sig") as f:
        f_csv = csv.writer(f)
        f_csv.writerow(headers)
        f_csv.writerows(data_list)

csv_path = r"/Users/loaferzk/LoaferZK/实习/深至/data/data_info.csv"
dump_to_csv(csv_path,data_list)
```

```python
source_dir = r"/Users/loaferzk/LoaferZK/实习/深至/data/测试集"
targetImg_dir = r"/Users/loaferzk/LoaferZK/实习/深至/data/测试集/image"
targetMask_dir = r"/Users/loaferzk/LoaferZK/实习/深至/data/测试集/masks"
if os.path.exists(targetImg_dir):
    pass
else:
    os.mkdir(targetImg_dir)
if os.path.exists(targetMask_dir):
    pass
else:
    os.mkdir(targetMask_dir)
    
for item in os.listdir(source_dir):
    imgPath = source_dir+"/"+item+"/"+"image"
    masksPath = source_dir+"/"+item+"/"+"masks"
    try:
        for item1 in os.listdir(imgPath):
            shutil.move(os.path.join(imgPath,item1),os.path.join(targetImg_dir,item1))
        for item2 in os.listdir(masksPath):
            shutil.move(os.path.join(masksPath,item2),os.path.join(targetMask_dir,item2))
    except FileNotFoundError as e:
        print("文件夹"+imgPath+"中不存在")
    continue
```
